```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sample = np.random.random(1000)
print("Shape:")
print(np.shape(sample))
print(f"Min = {np.min(sample)}")
print(f"Max = {np.max(sample)}")
print(f"Mean = {np.mean(sample)}")
print(f"Standard Deviation = {np.std(sample)}")
plt.hist(sample)
plt.show()
```

Distribution of Random Generated Numbers appears to be fairly uniform.

```python
sample = np.random.normal(0, 1, 1000)
print("Shape:")
print(np.shape(sample))
print(f"Min = {np.min(sample)}")
print(f"Max = {np.max(sample)}")
print(f"Mean = {np.mean(sample)}")
print(f"Standard Deviation = {np.std(sample)}")
plt.hist(sample)
plt.show()
```

I believe that the parameters for random.normal are:

random.normal(median/mean, standard deviation, sample size)

```python
sample = np.random.normal(1, .2, 1000000)
print("Shape:")
print(np.shape(sample))
print(f"Min = {np.min(sample)}")
print(f"Max = {np.max(sample)}")
print(f"Mean = {np.mean(sample)}")
print(f"Standard Deviation = {np.std(sample)}")
plt.hist(sample, bins=100)
plt.show()
```
