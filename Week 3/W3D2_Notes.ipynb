```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([2, 4, 3, 5, 0])

plt.plot(x, y)

plt.xticks([1.5, 4], ['tick 1', 'tick 2'])
plt.yticks([0.1, 0.5, 0.9, 2, 3, 4, 5])

plt.title('My Plot')
plt.xlabel('New X Label')
plt.ylabel('New Y Label')

plt.show()
```

```python
# Creates linear spacing // equal distribution of values between 1 and 5 with
x = np.linspace(1, 5, 100)
y_1 = np.log(x)
y_2 = x**2
y_3 = 2 ** x
plt.plot(x, x, 'b--', label='linear')
plt.plot(x, y_1, 'g:', label='log')
plt.plot(x, y_2, 'r-.', label='quadratic')
plt.plot(x, y_3, 'y.', label='exponential')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Runtime')
plt.show()
```
